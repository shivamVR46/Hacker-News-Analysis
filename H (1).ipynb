```
data = pd.read_csv("hn2018_2019.csv",
                   index_col=["Created At"], 
                   usecols = ["Title", "Post Type", "Created At"],
                   parse_dates = ["Created At"])

filteredData2018 = data.loc[data.index < '2019-01-01 00:00:00']
filteredData2018 = filteredData2018[0:1000]
regex = re.compile('[%s]' % re.escape(string.punctuation))
final_tokens = []
def Generate_Tokens(TEXT):
        # TEXT = TEXT.replace('\n',' ')
        Tokens = TEXT.lower().split()
        temp = []
        for words in Tokens:
             if words == '' or words in string.punctuation:
                 continue
             
             temp1  = re.split(r'\s+|[,-]\s*', words)
             temp.extend(temp1)
        
        for word in temp:
            word = regex.sub('', word)
            if word.isnumeric() or word == '':
                continue

            if word.isalnum() and len(word)>0:
                final_tokens.append(word)

        return final_tokens
for i in range(len(filteredData2018)):
  Generate_Tokens(data['Title'][i])
```

```

freqMap = {}
classFreq = {}

for index, row in filteredData2018.iterrows():
  tokens = Generate_Tokens(row['Title'])
  className = row['Post Type']
  
  if className in freqMap.keys():
      classFreq[className] += 1
  else:
      classFreq[className] = 1

  for word in tokens:
    appendClassName = word + "_" + className 
    if appendClassName in freqMap.keys():
      freqMap[appendClassName] += 1
    else:
      freqMap[appendClassName] = 1
```
